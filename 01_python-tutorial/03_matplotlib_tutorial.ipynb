### 기본 그래프 그리기
- `plot()` 함수: 선 그래프(Line graph) 그리기
- 축이름 설정: `title()`
- 타이틀 설정: `xlabel()`, `ylabel()`

````python
x = [1, 3, 5]
y = [2, 8, 11]

plt.plot(x, y)
plt.title('Graph')
plt.xlabel('X', loc='right')
plt.ylabel('Y')
plt.show()
````

````python
plt.rcParams.get('figure.figsize')
````

- 여러 개의 그래프 그리기

````python
fig = plt.figure() # figure
ax1 = fig.add_subplot(2, 2, 1) # axes: 행의 개수, 열의 개수, 첫번째
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(x, y)
ax1.set_title('Graph')

ax4.plot(x, y)

plt.show()
````

### 스타일 지정하기 (포맷문자열)
**포맷문자열**<br>
라인 그래프의 *색상, 스타일, 마커의 모양*을 문자로 지정해서 설정
```
fmt="[color][marker][line]"
```
**색상, 마커, 스타일 지정**<br>
- 색상 지정: [color](https://matplotlib.org/stable/gallery/color/named_colors.html)
- 선 스타일 지정: [linestyle](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)
- 점 지정: [marker](https://matplotlib.org/stable/api/markers_api.html)

|선 색상|선 스타일|마커 모양|
|:--:|:--:|:--:|
|'b': Blue|'-': Solid line|'.': Point|
|'g': Green|'--': Dashed line|'o': Circle|
|'r': Red|'-.': Dash-dot line|'^': Triangle|
|'k': Black|':': Dotted line|'*': star|

**linewidth 속성**<br>
plot() 함수의 linewidth 속성을 설정해서 라인의 두께를 설정

````python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, 'b--', marker='*')
# plt.plot(x, y, color='#00ff00', linewidth=5, linestyle='--')
plt.plot(x, z, color='navy')
plt.show()
````

### 그래프 형상(Figure) 설정
**범례 표시**<br>
plot() 함수의 label 파라미터 설정 및 plt.legend() 함수 호출
- 범례 표시하기: [legend](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

**x, y축의 범위 설정**<br>
- x축이 표시되는 범위 지정: `xlim()`
- y축이 표시되는 범위 지정: `ylim()`

**ticks 설정**<br>
- 그래프 x축의 눈금 표시: `xticks()`
- 그래프 y축의 눈금 표시: `yticks()`

**색상 설정 방법**<br>
- Base Color: 약자 사용 (예) : 'b', 'g', 'r' 등
- CSS Color: 색상의 이름 사용 (예)'navy'
  - 16진수 색상 코드 사용 (예) '#CC3333'

````python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, label='Eng')
plt.plot(x, z, label='Math')
plt.legend(loc='lower right')
plt.show()
````

### 산점도 (Scatter Plot)
- 데이터가 어느 정도 퍼져있는지(퍼진 정도)
- 머신러닝(ML)에서 자주 사용하는 그래프

````python
import gdown
import pandas as pd
````

````python
# !gdown https://raw.githubusercontent.com/devdio/datasets/main/iris.csv
````

````python
df = pd.read_csv('iris.csv', index_col=0)
df.shape
````

````python
df.head()
````

````python
plt.scatter(x=df['SepalLengthCm'], 
            y=df['SepalWidthCm'], color='r')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('iris')
plt.show()
````

````python
df['Species'].unique()
````

````python
df['Species'].value_counts()
````

- 각 종류('Species')별 꽃받침의 길이('SepalLengthCm')와 꽃받침의 너비('SepalWidthCm') 두 속성에 대한 산점도 (Scatter Plot) 그리기

````python
df_setosa = df[df['Species']=='Iris-setosa']
df_versicolor = df[df['Species']=='Iris-versicolor']
df_virginica = df[df['Species']=='Iris-virginica']

plt.scatter(x=df_setosa['SepalLengthCm'], 
            y=df_setosa['SepalWidthCm'], color='r')
plt.scatter(x=df_versicolor['SepalLengthCm'], 
            y=df_versicolor['SepalWidthCm'], color='g')
plt.scatter(x=df_virginica['SepalLengthCm'], 
            y=df_virginica['SepalWidthCm'], color='b')

plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('iris')
plt.show()
````

````python
plt.scatter(x=df.loc[df['Species']=='Iris-setosa', 'SepalLengthCm'], 
            y=df.loc[df['Species']=='Iris-setosa', 'SepalWidthCm'], color='r')
plt.scatter(x=df.loc[df['Species']=='Iris-versicolor', 'SepalLengthCm'], 
            y=df.loc[df['Species']=='Iris-versicolor', 'SepalWidthCm'], color='g')
plt.scatter(x=df.loc[df['Species']=='Iris-virginica', 'SepalLengthCm'], 
            y=df.loc[df['Species']=='Iris-virginica', 'SepalWidthCm'], color='b')

plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('iris')
plt.show()
````

- 각 종류('Species')별 꽃잎의 길이('PetalLengthCm')와 꽃잎의 너비('PetalWidthCm') 두 속성에 대한 산점도 (Scatter Plot) 그리기

````python
df_setosa = df[df['Species']=='Iris-setosa']
df_versicolor = df[df['Species']=='Iris-versicolor']
df_virginica = df[df['Species']=='Iris-virginica']
	
plt.scatter(x=df_setosa['PetalLengthCm'], 
            y=df_setosa['PetalWidthCm'], color='r')
plt.scatter(x=df_versicolor['PetalLengthCm'], 
            y=df_versicolor['PetalWidthCm'], color='g')
plt.scatter(x=df_virginica['PetalLengthCm'], 
            y=df_virginica['PetalWidthCm'], color='b')

plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.title('iris')
plt.show()
````

````python
plt.scatter(x=df.loc[df['Species']=='Iris-setosa', 'PetalLengthCm'], 
            y=df.loc[df['Species']=='Iris-setosa', 'PetalWidthCm'], color='r')
plt.scatter(x=df.loc[df['Species']=='Iris-versicolor', 'PetalLengthCm'], 
            y=df.loc[df['Species']=='Iris-versicolor', 'PetalWidthCm'], color='g')
plt.scatter(x=df.loc[df['Species']=='Iris-virginica', 'PetalLengthCm'], 
            y=df.loc[df['Species']=='Iris-virginica', 'PetalWidthCm'], color='b')

plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.title('iris')
plt.show()
````

- 여러 개의 그래프 그리기

````python
fig = plt.figure(figsize=(10, 4))

ax1 = fig.add_subplot(1, 2, 1)
ax1.scatter(x=df.loc[df['Species']=='Iris-setosa', 'SepalLengthCm'], 
            y=df.loc[df['Species']=='Iris-setosa', 'SepalWidthCm'], color='r')
ax1.scatter(x=df.loc[df['Species']=='Iris-versicolor', 'SepalLengthCm'], 
            y=df.loc[df['Species']=='Iris-versicolor', 'SepalWidthCm'], color='g')
ax1.scatter(x=df.loc[df['Species']=='Iris-virginica', 'SepalLengthCm'], 
            y=df.loc[df['Species']=='Iris-virginica', 'SepalWidthCm'], color='b')
ax1.set_xlabel('SepalLengthCm')
ax1.set_ylabel('SepalWidthCm')
ax1.set_title('Sepal')

ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(x=df.loc[df['Species']=='Iris-setosa', 'PetalLengthCm'], 
            y=df.loc[df['Species']=='Iris-setosa', 'PetalWidthCm'], color='r')
ax2.scatter(x=df.loc[df['Species']=='Iris-versicolor', 'PetalLengthCm'], 
            y=df.loc[df['Species']=='Iris-versicolor', 'PetalWidthCm'], color='g')
ax2.scatter(x=df.loc[df['Species']=='Iris-virginica', 'PetalLengthCm'], 
            y=df.loc[df['Species']=='Iris-virginica', 'PetalWidthCm'], color='b')
ax2.set_xlabel('PetalLengthCm')
ax2.set_ylabel('PetalWidthCm')
ax2.set_title('Petal')

plt.suptitle('iris scatter plot', fontsize=15)
plt.tight_layout()
plt.show()
````

- 여러 개의 그래프 그리기: `subplots(row, column)` 함수
  - 리턴되는 Axes 객체 배열을 사용해서 그래프 그리기

````python
mydata = [[1, 2, 3, 4], 
          [10, 11, 12, 13], 
          [30, 31, 32, 33], 
          [41, 43, 45, 47]]

fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(mydata[0])
ax[0, 1].plot(mydata[1])
ax[1, 0].plot(mydata[2])
ax[1, 1].plot(mydata[3])

plt.show()
````

**이미지 데이터: 3차원**

````python
plt.figure(figsize=(5, 5))
plt.imshow(arr)
plt.xticks([])
plt.yticks([])
plt.title('puppy')
plt.show()
````

- 여러 개의 이미지

````python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

fig_name = ['image1', 'image2', 'image3']
axes = ax.reshape(-1)

for i, axis in enumerate(axes):
    img = Image.open('img/'+fig_name[i]+'.jpg')
    arr = np.array(img)
    axis.imshow(arr)
    axis.set_xticks([])
    axis.set_yticks([])
````
