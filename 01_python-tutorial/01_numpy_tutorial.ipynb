### 다차원 배열
- `type() 함수`: 객체의 데이터 타입을 반환해주는 함수

### 배열의 속성
|속성|설명|
|:--:|:--:|
|shape|배열의 형태를 튜플 자료형으로 표현|
|ndim|배열의 차원(축의 개수)|
|size|배열의 원소 개수|
|dtype|배열 원소의 자료형(모든 원소가 같은 자료형)|
|itemsize|각 배열 원소의 바이트의 크기|

### 배열의 자료형(Data Type)
- 불리언(Boolean)
- 정수(Integer)
- 부호 없는 정수(Unsigned Integer)
- 부동소수점(Float)
- 복소 부동소수점(Complex) 등

`dtype`: 자료형 확인, `astype`: 자료형 변환

1-D Array

```python
john = [10, 8, 6, 7]
x = np.array(john)

print(x)
print(type(x))
```

2-D Array

```python
scores = [[1, 2, 3, 4], 
          [5, 6, 7, 8]]
x = np.array(scores)

print(x)
print(x.shape)
print(x.dtype)
```

```python
def print_arr(arr):
    print(arr)
    print('shape:', arr.shape)
    print('ndim:', arr.ndim)
    print('size:', arr.size)
    print('dtype:', arr.dtype)
    print('itemsize:', arr.itemsize)
```

```python
x = np.array([1., 2., 3., 4., 5.])
print_arr(x)
```

```python
y = x.astype(np.int32)
print_arr(y)
```

### 배열의 형태(Shape)와 축(Axis)
- 배열의 형태(Shape): (axis 0 원소 개수, axis 1 원소 개수, axis 2 원소 개수)
- 배열의 축(Axis)
  - 1-D array: 축 1개(axis 0)
  - 2-D array: 축 2개(axis 0, axis 1)
  - 3-D array: 축 3개(axis 0, axis 1, axis 2)

### 배열 생성함수
- `arange() 함수`: np.arange(시작, 끝, 개수)으로 끝을 포함하지 않고, 개수는 간격을 의미함
- `linspace() 함수`: np.linspace(시작, 끝, 개수)으로 시작과 끝을 포함하고, 개수는 해당 구간의 개수를 의미함
- `zeros() 함수`: 주어진 형태의 배열에 모두 0으로 채움
- `ones() 함수`: 주어진 형태의 배열에 모두 1로 채움
- `full() 함수`: 주어진 형태의 배열에 모두 주어진 값으로 채움
- `ones_like(), zeros_like() 함수`: 주어진 형태와 타입을 갖는 배열에 모두 0으로 채움, 주어진 형태와 타입을 갖는 배열에 모두 1로 채움
- `empyt()`, `empty_like()`, `np.eye()` 등 함수 제공

```python
x = np.array([1, 2, 3])
x
```

```python
x + y
```

### 배열 연산
- `np.add()`
- `np.subtract()`
- `np.divide()`
- `np.multiply()`
- `np.dot()`
- `np.matmul()`

```python
x = np.array([[1, 2], 
             [3, 4]])
y = np.array([[5, 6], 
              [7, 8]])
```

```python
x + y
```

```python
np.add(x, y)
```

```python
x - y, np.subtract(x, y)
```

```python
x / y, np.divide(x, y)
```

```python
x * y, np.multiply(x, y)
```

`np.dot()`, `np.matmul()`
- 2차원 계산은 동일함
- 3차원 이상은 계산 방식이 달라짐

```python
# 행렬 곱
x @ y, np.dot(x, y)
```

```python
np.matmul(x, y)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

np.matmul(x, y)
```

```python
x = np.array([[1, 2], 
             [3, 4]])
y = np.array([[5, 6], 
              [7, 8]])

np.matmul(x, y)
```

### 배열 저장: savez()
- 넘파이 배열을 여러 개 저장할 수 있음

```python
x = np.random.randn(5, 4)
y = np.random.randn(3, 3)
```

```python
x
```

```python
y
```

```python
# 파일 저장
np.savez('mydata.npz', xvar=x, yvar=y)
```

```python
# 파일 로딩
mydata = np.load('mydata.npz')
```

```python
x = mydata['xvar']
x
```

```python
x = mydata['yvar']
x
```
