# 선형 회귀(Linear Regression)
---

### 필요한 라이브러리 불러오기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

### 데이터 준비

```python
# 3x + 6
X_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24.])
```

### 모델(Model)
1. Sequential API
2. Functional API
3. Subclassing API

### 모델링(Modeling)
- **Sequential()**
- **Dense**: 완전 연결층
  - `units`:출력 뉴런의 수
  - `input_shape`: 입력 뉴런의 수
  - `activation`: 활성화 함수
- **summary()**: 모델의 정보 요약

```python
model = keras.Sequential([
    layers.Dense(units=1, input_shape=(1,))
])
```

```python
model.summary()
```

### 컴파일(Compile)
- **compile()**: 모델 컴파일(옵티마이저, 손실함수,메트릭 함수 설정)
  - `optimizer`: 학습을 위한 옵티마저 설정
  - `loss`: 학습 때 사용하는 손실함수 설정
  - `metrics`: 학습 모니터링 지표

```python
model.compile(
    optimizer='sgd', 
    loss='mse', 
    metrics=['mse', 'mae']
)
```

### 학습(Training)
- **fit()**: 모델 학습
  - `train feature`
  - `train label`
  - `epochs`: 에포크
  - `batch_size`: 베치 크기
  - `validation_data=(X_val, y_val)`: 검증용 데이터
  - `validation_split`: 검증용 데이터의 비율
  - `verbose`: 학습할 때 출력되는 문구(0, 1, 2)

```python
model.fit(X_train, y_train, 
          epochs=1000, 
          batch_size=len(X_train), 
          verbose=1)
```

```python
model.get_weights()
```

### 평가(Evaluation)

```python
model.evaluate(X_train, y_train)
```

### 예측(Prediction)

```python
X_test = [10.]

y_pred = model.predict(X_test)
y_pred
```
